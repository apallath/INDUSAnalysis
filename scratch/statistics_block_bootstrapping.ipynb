## Autocorrelation analysis (Sec. 7.3.1, Grossfield et al)

```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import pymbar
from pymbar import testsystems

from statsmodels.graphics import tsaplots

import numba
from numba import jit
```

```python
mpl.rcParams['figure.dpi'] = 100
```

```python
A_t = testsystems.correlated_timeseries_example(N=1000, tau=5.0)
plt.plot(A_t)
```

```python
plt.figure()
tsaplots.plot_acf(A_t); #semicolon suppresses output (double graphs)
ax = plt.gca()
ax.set_xlabel("Lags, $j$")
ax.set_ylabel("ACF, $C_j$")
```
