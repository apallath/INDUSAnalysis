```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pymbar
import statsmodels.tsa.stattools
```

```python
#load time series data of waters in a probe volume generated by GROMACS-INDUS
ts = np.load("test-4.25_N.npy")
t = ts[0,:]
x = ts[1,:]
tstep = t[1] - t[0]
print(tstep)
```

```python
plt.plot(t,x)
```

# Equilibrated region

```python
start = int(np.floor(600/tstep))
plt.plot(t[start:],x[start:])
```

# ACF

```python
acf = statsmodels.tsa.stattools.acf(x[start:],nlags=100)
```

```python
plt.plot(acf)
```
