```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pymbar
from pymbar import testsystems
import statsmodels.tsa.stattools
from scipy import stats
```

```python
#load time series data of waters in a probe volume generated by GROMACS-INDUS
ts = np.load("test-4.25_N.npy")
t = ts[0,:]
x = ts[1,:]
tstep = t[1] - t[0]
print(tstep)
```

```python
plt.plot(t,x)
```

# Equilibrated region

```python
start = int(np.floor(600/tstep))
plt.plot(t[start:],x[start:])
```

# ACF

```python
A_t = testsystems.correlated_timeseries_example(N=1000, tau=5.0)
acf = statsmodels.tsa.stattools.acf(A_t,fft=True)
T_acf = np.array(range(len(acf)))
```

```python
plt.figure()
plt.plot(A_t)
plt.figure()
plt.plot(T_acf, acf)
```

```python
#Fit exponential to ACF
tau = pymbar.timeseries.integratedAutocorrelationTime(A_t)
print(tau)
plt.plot(np.exp(-T_acf/tau))
plt.plot(acf)
```

```python
#ACF for our data
xe = x[start:]
te = t[start:]
acf = statsmodels.tsa.stattools.acf(xe,fft=True,nlags=200)
plt.plot(range(len(acf)), acf)
tau_step = pymbar.timeseries.integratedAutocorrelationTime(xe)
print(tau_step)
#effective number of uncorrelated samples
N_uncorr = len(xe)/tau_step
print()
print(N_uncorr)
```

```python
sidx = np.random.choice(np.array(range(len(xe))), size = int(N_uncorr), replace=False)
print(sidx)
```

```python
plt.scatter(te[sidx], xe[sidx])

#Correlated data
#Standard deviation
print(np.std(xe))

#Standard error of the mean
print(stats.sem(xe[sidx]))

#Decorrelated samples
std = 0
sem = 0
for i in range 1000:
    std += np.std(xe[sidx]))
    print(stats.sem(xe))
```
